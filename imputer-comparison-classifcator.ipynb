```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
imputers = ['KNN Imputer', 'Simple Imputer', 'Iterative Imputer']
models = ['DecisionTree', 'RandomForest', 'Bagging', 'GradientBoosting']
accuracy = [[0.5105, 0.5777, 0.5867],  # DecisionTree
            [0.6429, 0.6830, 0.6999],  # RandomForest
            [0.6310, 0.6667, 0.6563],  # Bagging
            [0.6297, 0.6679, 0.7260]]  # GradientBoosting

f1_score = [
    [0.5159, 0.5736, 0.5906],  # DecisionTree
    [0.6231, 0.6651, 0.6669],  # RandomForest
    [0.6150, 0.6510, 0.6376],  # Bagging
    [0.6149, 0.6571, 0.7083],  # GradientBoosting
]

precision = [
    [0.5168, 0.5720, 0.5969],  # DecisionTree
    [0.6321, 0.6822, 0.6605],  # RandomForest
    [0.6303, 0.6616, 0.6323],  # Bagging
    [0.6200, 0.6649, 0.6982],  # GradientBoosting
]

recall = [
    [0.5150, 0.5777, 0.5867],  # DecisionTree
    [0.6429, 0.6830, 0.6999],  # RandomForest
    [0.6310, 0.6667, 0.6563],  # Bagging
    [0.6297, 0.6679, 0.7260],  # GradientBoosting
]
```

```python
def plotting_evaluation(metric_name, values):
    # Plot settings
    bar_width = 0.2
    x = np.arange(len(models))

    # Plot imputers
    plt.figure(figsize=(10, 6))
    for i, imputer in enumerate(imputers):
        plt.bar(x + i * bar_width, [value[i] for value in values], width=bar_width, label=imputer)

    plt.xlabel('Models')
    plt.ylabel(metric_name)
    plt.title(f'{metric_name} for Different Models and Imputers')
    plt.xticks(x + bar_width, models)
    plt.legend()
    plt.show()
```

```python
plotting_evaluation('accuracy', accuracy)
plotting_evaluation('F1-score', f1_score)
plotting_evaluation('Precision', precision)
plotting_evaluation('Recall', recall)
```
