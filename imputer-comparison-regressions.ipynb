```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data (example structure based on your table)
models = ['DecisionTree', 'RandomForest', 'Bagging', 'Linear', 'SVM']
imputers = ['KNN Imputer', 'Simple Imputer', 'Iterative Imputer']

# Mean Squared Error
mse_values = [
    [2.8011, 2.3443, 1.7743],  # DecisionTree
    [1.4142, 1.2135, 0.9198],  # RandomForest
    [1.4106, 1.2485, 1.0598],  # Bagging
    [3.2139, 3.487, 3.73],     # Linear
    [2.8378, 3.1402, 3.0174]   # SVM
]

# Mean Absolute Error
mae_values = [
    [0.9645, 0.7488, 0.7562],  # DecisionTree
    [0.8156, 0.689, 0.6551],   # RandomForest
    [0.7983, 0.6918, 0.691],   # Bagging
    [1.4741, 1.5602, 1.6465],  # Linear
    [1.2448, 1.2868, 1.2709]   # SVM
]

# R Score
r2_values = [
    [0.2738, 0.4031, 0.5478],  # DecisionTree
    [0.6333, 0.691, 0.7659],   # RandomForest
    [0.6343, 0.6821, 0.7299],  # Bagging
    [0.1668, 0.1121, 0.0493],  # Linear
    [0.2643, 0.2004, 0.2309]   # SVM
]
```

```python
#Function that plots the values
def plotting_evaluation(metric_name, values):
    # Plot settings
    bar_width = 0.2
    x = np.arange(len(models))

    # Plot Mean Squared Error
    plt.figure(figsize=(10, 6))
    for i, imputer in enumerate(imputers):
        plt.bar(x + i * bar_width, [value[i] for value in values], width=bar_width, label=imputer)

    plt.xlabel('Models')
    plt.ylabel(metric_name)
    plt.title(f'{metric_name} for Different Models and Imputers')
    plt.xticks(x + bar_width, models)
    plt.legend()
    plt.show()
```

```python
plotting_evaluation('Mean Squared Error', mse_values)
plotting_evaluation('Mean Absolute Error', mae_values)
plotting_evaluation('R Score', r2_values)
```
