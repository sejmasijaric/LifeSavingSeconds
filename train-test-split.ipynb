```python
final_dataset = pd.concat([troponin_dataset, troponin_dataset_whole], ignore_index=True)
print(final_dataset.head(10))
```

```python
#Columns that are not relevant or double in the dataset
columns_to_eliminate = ['WBCValue', 'TNCValue', 'PLTIValue', 'AFNR','ttt',  'AFNR_kopieraster', 'AFNR_doppelt', 'ttt', 'Kommentar', 'AANR', 'Gender', 'Geb', 'PatientBirthday', 'Geburtstag_komb', 'Geschlecht_komb',  'Trop_2Kat', 'TROPiHSMesswert', 'dattag', 'Gender', 'Sex', 'Abnahme', 'Eingang', 't', 'Fertig', 'MeasurementDateandTime' ]
final_dataset.drop(columns_to_eliminate, axis=1, inplace=True)
final_dataset.head(10)
```

```python
# Define the threshold for missing values (60%)
threshold = 0.8

# Calculate the percentage of missing values for each column
missing_percentage = final_dataset.isnull().mean()

# Drop columns where the percentage of missing values is greater than the threshold
final_dataset = final_dataset.loc[:, missing_percentage <= threshold]

# Display the remaining columns
print("Columns after removing those with more than 80% missing values:")
print(f"Number of columns: {len(final_dataset.columns)}")
print(final_dataset.columns)
```

```python
# Set boundaries for bins
lower_boundary = 15
upper_boundary = 50

# Define number of bins
bins_below_15 = 2
bins_between = 3
bins_above_50 = 2

edges_below = np.linspace(final_dataset['TROPTHSMesswert'].min(), lower_boundary, bins_below_15 + 1)
edges_between = np.linspace(lower_boundary, upper_boundary, bins_between + 1)
edges_above = np.linspace(upper_boundary, final_dataset['TROPTHSMesswert'].max(), bins_above_50)

edges = np.unique(np.concatenate([edges_below, edges_between, edges_above]))

final_dataset['TROPTHSMesswert'] = pd.cut(final_dataset['TROPTHSMesswert'], bins=edges, labels=False)

print(final_dataset['TROPTHSMesswert'])
```

```python
# Print the number of unique labels (bins)
num_labels = final_dataset['TROPTHSMesswert'].nunique()
print(f"Number of labels (bins): {num_labels}")

# Print the count of cases in each label
label_counts = final_dataset['TROPTHSMesswert'].value_counts().sort_index()
print("\nNumber of cases in each label:")
print(label_counts)
```

```python
# replace non-numerical values with missing values; imputation is not possible with them
# values like CLOT, MUEBS, ',,,', hm, etc.
for column in final_dataset.columns:
    final_dataset[column] = pd.to_numeric(final_dataset[column], errors='coerce')

# Define splitting boundaries
split_1 = int(0.8 * len(final_dataset))
split_2 = int(0.9 * len(final_dataset))

# Shuffle dataset and reset index
final_dataset = final_dataset.sample(n = len(final_dataset))
final_dataset = final_dataset.reset_index(drop=True)

# Split dataset in train/test/split
train_set = final_dataset[: split_1]
dev_set = final_dataset[split_1:split_2]
test_set = final_dataset[split_2:]
```

```python
# Impute missing values
imputer = IterativeImputer()

#Fit and transform sets
train_imputed = imputer.fit_transform(train_set)
test_imputed = imputer.transform(test_set)
dev_imputed = imputer.transform(dev_set)
```

```python
#Convert into DataFrame
pd_train = pd.DataFrame(train_imputed, columns=train_set.columns)
pd_test = pd.DataFrame(test_imputed, columns=test_set.columns)
pd_dev = pd.DataFrame(dev_imputed, columns=dev_set.columns)

whole_dataset = pd.concat([pd_train, pd_test, pd_dev], ignore_index=True)
```

```python
# should we shuffle the data before saving it to csv files
pd_train.to_csv('data/train-dataset.csv')
pd_dev.to_csv('data/dev-dataset.csv')
pd_test.to_csv('data/test-dataset.csv')
whole_dataset.to_csv('data/whole-dataset.csv')
```
